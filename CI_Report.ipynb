```python
New_df = pd.DataFrame(sales, columns= ['purchase_sum', 'experience'])
```

```python
New_Target = pd.DataFrame(sales.characteristic_one)
New_Target.columns = ['Targets']
```

```python
colormap = np.array(['','red','black'])
plt.scatter(New_df.purchase_sum, New_df.experience, c= colormap[New_Target.Targets])
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 2 )
```

```python
model.fit(New_df)
```

```python
model.labels_
```

```python
colormap = np.array(['red','black'])
plt.scatter(New_df.purchase_sum, New_df.experience, c=colormap[model.labels_])
```
